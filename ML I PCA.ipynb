```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data
```

```python
data = pd.read_csv("Resources/global_heat_index.csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# cleaning the data
clean1 = data[data['Hour']<6]
clean2 = data[data["Hour"]>18]

print("cleaning the data from 12:00 am to 6::00 am")
clean1.head()
```

```python
print("cleaning the data from 18:00 pm to 12::00 am")
clean2.head()
```

```python
data = data.drop(clean1.index, axis =0)
data = data.drop(clean2.index, axis =0)
```

```python
print("Cleaned data description: ")
data.describe()
```

```python
print("Cleaned data: ")
data.head()
```

```python
# preparing training and testing data
```

```python
x = data.iloc[0: , 4:10].values
y = data.iloc[:, -1].values

y = y.reshape(-1,1)

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
# Standardising data
```

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()
xtrain = sc_x.fit_transform(xtrain)
xtest = sc_x.transform(xtest)

sc_y = StandardScaler()
ytrain = sc_y.fit_transform(ytrain)
```

```python
print(xtrain)
```

```python
print(ytrain)
```

```python
# Kmeans with 6 clusters
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=6, random_state=0)
km = km.fit(xtrain)
```

```python
# view predict class
labels = km.labels_
print("labels: ", labels)

# view clusters centers
cluster_centers = km.cluster_centers_
print(" \n\ncluster_centers:\n", cluster_centers)
```

```python
# we cant visualise this 6D data, so we will use PCA to reduce the dimesions and to visualise our data
```

```python
# Apply PCA - Dimensionality Reduction
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2).fit(xtrain).transform(xtrain)

pca
```

```python
# Visualisation
```

```python
for i in range (0, pca.shape[0]):
    if km.labels_[i]== 0:
        c1_6 = plt.scatter(pca[i, 0], pca[i, 1], marker='^', c= "orange")
    elif km.labels_[i]== 1:
        c2_6 = plt.scatter(pca[i, 0], pca[i, 1], marker='*', c= "red")
    elif km.labels_[i]== 2:
        c3_6 = plt.scatter(pca[i, 0], pca[i, 1], marker='+', c= "yellow")
    elif km.labels_[i]== 3:
        c4_6 = plt.scatter(pca[i, 0], pca[i, 1], marker='o', c= "green")
    elif km.labels_[i]== 4:
        c5_6 = plt.scatter(pca[i, 0], pca[i, 1], marker='<', c= "b")
    elif km.labels_[i]== 5:
        c6_6 = plt.scatter(pca[i, 0], pca[i, 1], marker='>', c= "cyan")
    
plt.rcParams["figure.figsize"]= (12,10)
plt.legend([c1_6,c2_6,c3_6,c4_6,c5_6,c6_6], ["Cluster 0", "Cluster 1"
                                             , "Cluster 2","Cluster 3"
                                             , "Cluster 4", "Cluster 5"])
plt.title("K means clustering with 6 clusters")
plt.show()
```
